```python
import pandas as pd                        
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='pt-br', tz=180)
```

```python
kw_list = ['Removedor','removedor espaço','removedor caracter']
df = pytrends.get_historical_interest(kw_list, year_start=2022, month_start=1, day_start=1, hour_start=0, year_end=2022, month_end=5, day_end=16, hour_end=20, cat=0, geo='', gprop='', sleep=0)
```

```python
lista = ['Futebol']
kw_list = lista[0]
df = pytrends.get_historical_interest(lista, year_start=2022, month_start=1, day_start=1, hour_start=0, year_end=2022, month_end=5, day_end=16, hour_end=20, cat=0, geo='', gprop='', sleep=0)
df = pd.DataFrame(df[kw_list])
df[f'{kw_list}']= df[f'{kw_list}'].map(str)
mask0 =df[f'{kw_list}'].str.contains(r'(?:\s|^)0(?:\s|$)') == False
df = df.loc[mask0] 
df.sort_values(by=[f'{kw_list}'], ascending=False)
```

```python
df = pd.DataFrame(df['Removedor'])
```

```python
df['Removedor']= df['Removedor'].map(str)
mask0 =df['Removedor'].str.contains(r'(?:\s|^)0(?:\s|$)') == False
df = df.loc[mask0] 
df.sort_values(by=['Removedor'], ascending=False)
```
